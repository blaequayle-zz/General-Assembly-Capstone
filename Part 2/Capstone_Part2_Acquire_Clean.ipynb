```python
def import_csv(csvlist, year, encoding):
    '''Imports csv file from list of csv and creates new dataframe which contains all the csv files'''
    cycle_df = pd.DataFrame()
    for csvfile in csvlist:
        file_name = './Capstone_data/' + str(year) + '/' + csvfile
        df = pd.read_csv(file_name, parse_dates = [3, 6], encoding = encoding, infer_datetime_format = True, dayfirst = True)
        cycle_df = cycle_df.append(df)
    print 'Shape of cycle data {}'.format(year),':', cycle_df.shape
    print 'Info on cycle data {}'.format(year),':', cycle_df.info()
    return cycle_df
```

### Web APIs: WorldWeatherOnline
####  Accessing API
First, obtaining weather data for central London (W1 postcode) from the **WorldWeatherOnline** API. The requests package is required. The following function carries out:
* Appends current access key, postcode of interest, the startdate and the enddate to the url string for the API. 
* A get request is then made to the API using the url and saved as `resp`. The status code of the API is printed to ensure there are no issues, code 200 should be displayed if this is the case. 
* The response from the API is assigned to `j` and returned by the function.

```python
def weather_api_data(postcode, startdate, enddate):
    '''Obtains data via API from World Weather Online'''
    
    # Define url of API
    key = str(weather_key)
    postcode = str(postcode)
    startdate = str(startdate)
    enddate = str(enddate)
    payload = {'key': key, 'q': postcode, 'format': 'json', 'date': startdate, 'enddate': enddate, 'includelocation': 'yes', 'tp': '6'}
    
    # Request information from API
    r = requests.get('http://api.worldweatheronline.com/premium/v1/past-weather.ashx', params=payload)

    # Check status code for API
    print r.status_code

    # Assign output in json format and return
    j = r.json()
    return j
```

#### Extracting Data
The output is in JSON format per day, and includes alot of unwanted data within nested dictionaries. The following function is used to extract the required data and transform it into a dataframe format. The workflow is as follows:
* Set up one list for every variable of interest in the JSON file
* Format start and end dates into datatime objects using `.strptime`
* Determine `day_total` i.e. the difference between the start and end date in days
* Use as input to the `range` function (plus 1 to account for indexing) to loop over the dictionaries (one per date) within the JSON 
* For each date (d), the appropriate list is appended with the correct variable. As an example: <br> `winddir16Point.append(jsondata["data"]["weather"][2]["hourly"][2]['winddir16Point'])` takes the JSON and looks inside "data" and "weather" dictionaries to find the date at index 2, then the "hourly" dictionary at index 2 which records the value of wind direction at 12 noon on that day under the `winddir16Point` key. <br>
* Lists are used to create a dataframe, the dataframe is not saved to csv as this function is used as part of a larger function.

```python
def extract_weather_data(jsondata, startdate, enddate):
    '''Accepts json data and extracts data of interest into dataframe'''
    
    # Set up lists to append data to
    date = []
    winddir16Point = [] 
    FeelsLikeC = []
    windspeedMph = []
    WindChillC = []
    WindGustMiles = []
    cloudcover = []
    humidity = []
    precipMM = []
    tempC = []
    description = [] 
    winddirDegree = []
    maxtempC = []
    mintempC = []
    sunHour = []
    totalSnow_cm = []
    uvIndex = []
    
    # Format start and end dates to determine interval to use
    start = datetime.strptime(startdate, '%Y-%m-%d')
    end = datetime.strptime(enddate, '%Y-%m-%d')
    days = end - start
    day_total = int(days.total_seconds() / timedelta (days=1).total_seconds())
    
    # Using 2 to capture actual weather at 12 noon
    for d in range(day_total+1):
        date.append(jsondata["data"]["weather"][d]["date"])
        winddir16Point.append(jsondata["data"]["weather"][d]["hourly"][2]['winddir16Point'])
        FeelsLikeC.append(jsondata["data"]["weather"][d]["hourly"][2]['FeelsLikeC'])
        windspeedMph.append(jsondata["data"]["weather"][d]["hourly"][2]['windspeedMiles'])
        WindChillC.append(jsondata["data"]["weather"][d]["hourly"][2]['WindChillC'])
        WindGustMiles.append(jsondata["data"]["weather"][d]["hourly"][2]['WindGustMiles'])
        cloudcover.append(jsondata["data"]["weather"][d]["hourly"][2]['cloudcover'])
        humidity.append(jsondata["data"]["weather"][d]["hourly"][2]['humidity'])
        precipMM.append(jsondata["data"]["weather"][d]["hourly"][2]['precipMM'])
        tempC.append(jsondata["data"]["weather"][d]["hourly"][2]['tempC'])
        description.append(jsondata["data"]["weather"][d]["hourly"][2]['weatherDesc'][0]['value'])
        winddirDegree.append(jsondata["data"]["weather"][d]["hourly"][2]['winddirDegree'])
        maxtempC.append(jsondata["data"]["weather"][d]['maxtempC'])
        mintempC.append(jsondata["data"]["weather"][d]['mintempC'])
        sunHour.append(jsondata["data"]["weather"][d]['sunHour'])
        totalSnow_cm.append(jsondata["data"]["weather"][d]['totalSnow_cm'])
        uvIndex.append(jsondata["data"]["weather"][d]['uvIndex'])
    
    weather = pd.DataFrame({ 'date': date, 'wind_dir_16pt': winddir16Point, 'feels_like' : FeelsLikeC, 'wind_speed': windspeedMph, 'wind_chill': WindChillC, 'wind_gust':  WindGustMiles, 'cloud_cover': cloudcover, 'humidity': humidity, 'precipitation': precipMM, 'temperature': tempC, 'description': description, 'wind_dir_deg' : winddirDegree, 'max_temp': maxtempC, 'min_temp': mintempC, 'sun_hours' : sunHour, 'total_snow' : totalSnow_cm, 'uv_index': uvIndex})
    
    return weather
```

#### Issue: Download Limit
The WorldWeatherOnline API limits downloads to 31 days at a time. A for loop is used to iterate over the start dates and then enter into each function - `weather_api_data` and `extract_weather_data`. The function uses i=30 as an arbitrary number of days in a month as the aim is to acquire all the data, it does not matter if it not exactly one month each time, as long as it is successive. The number used for range is chosen to be sufficiently large so that all the data is captured from 2012 until the present. Each additional block is appended onto the dataframe.

```python
def aggregate_weather_data(startdate):
    '''Accepts start date to begin acquiring data'''
    
    #Format the startdate input into datetime object
    start = datetime.strptime(startdate, '%Y-%m-%d')
    weather = pd.DataFrame()

    for _ in range(100):
        i = 30
        try:
            # Set up start date and end date for use in function
            start_date = str(start.date())
            end_date = str(start.date() + timedelta(days=i))
            
            # Call weather_api_data function
            json = weather_api_data('W1', start_date, end_date)
            
            # Call extract_weather_data function
            df = extract_weather_data(json, start_date, end_date)
            
            # Append output to the weather dataframe
            weather = weather.append(df)
            
            # Add on i days to the start date.
            start = start + timedelta(days=i)
        except:
            pass
    
    weather.to_csv('historical_weather.csv', encoding='utf-8')
```

### Web APIs:  Transport for London
#### Accessing API & Extracting Data
The stages in the first two functions above have been combined for brevity. The time zone for the API is GMT or BST as appropriate.

```python
def dock_api_data():
    '''Obtains data via Unified API from TFL'''
    
    key_id = str(key_id)
    app_id = str(app_id)

    # Define url of API
    url = 'https://api.tfl.gov.uk/bikepoint?app_key=' + key_id + '&app_id=' + app_id
    
    # Request information from API
    resp = requests.get(url)
    print resp.status_code
    j = resp.json()
    
    # Set up lists to append data to
    Location = []
    Latitude = [] 
    Longitude = []
    Terminal_ID = []
    Installed = []
    Locked = []
    Install_Date = []
    Removal_Date = []
    Temporary = []
    Number_Bikes = []
    Number_Empty = [] 
    Number_Docks = []

    
    for i, chunk in enumerate(j):
        Location.append(j[i]['commonName']) # Locality
        Latitude.append(j[i]['lat']) # Latitude
        Longitude.append(j[i]['lon']) # Longitude
        Terminal_ID.append(j[i]['additionalProperties'][0]['value']) # TerminalName
        Installed.append(j[i]['additionalProperties'][1]['value']) # Installed
        Locked.append(j[i]['additionalProperties'][2]['value']) # Locked
        Install_Date.append(j[i]['additionalProperties'][3]['value']) # Install Date
        Removal_Date.append(j[i]['additionalProperties'][4]['value']) # Removal Date
        Temporary.append(j[i]['additionalProperties'][5]['value']) # Temporary
        Number_Bikes.append(j[i]['additionalProperties'][6]['value']) # Number Bikes
        Number_Empty.append(j[i]['additionalProperties'][7]['value']) # Number Empty Docks
        Number_Docks.append(j[i]['additionalProperties'][8]['value']) # Number Docks
    
    docks = pd.DataFrame({ 'locality': Location, 'lat': Latitude, 'lon' : Longitude, 'id': Terminal_ID, 'installed': Installed, 'locked':  Locked, 'install_date': Install_Date, 'removal_date': Removal_Date, 'temporary': Temporary, 'num_bikes': Number_Bikes, 'num_empty': Number_Empty , 'num_docks': Number_Docks})
    
    docks.to_csv('bike_share_docks.csv', encoding='utf-8')
    
    return docks
```

`install_date` and `removal_date` columns are converted from unix timestamp format into datetime64 using the defined function. There are 774 docks documented in total.
```python
def convert_unix_dates(column):
    '''Takes column from dataframe and converts values from Unix timestamp
    format to string in %Y-%m-%d %H:%M:%S format'''
    dates = []
    for x in column:
        try:
            r = int(x/1000)
            d = datetime.fromtimestamp(int(r)).strftime('%Y-%m-%d %H:%M:%S')
            dates.append(d)
        except:
            dates.append(x)
    return dates
```

```python
# Apply function above to convert dates into recognisable format before setting as datetime format
docks['install_date'] = convert_unix_dates(docks.install_date)
docks['removal_date'] = convert_unix_dates(docks.removal_date)
docks['install_date'] =  pd.to_datetime(docks['install_date'])
docks['removal_date'] =  pd.to_datetime(docks['removal_date'])
```

###Melt: Classification dataset
Creating a further dataset for use in classification task - predicting whether a dock will be full, have spaces or it is not possible to determin for a certain dock for a specific hour.

```python
X2016_geo = pd.read_csv('X2016_geo.csv', index_col = 0)
X2016_geo['start_date'] =  pd.to_datetime(X2016_geo['start_date'], format='%Y-%m-%d %H:%M:%S')
X2016_geo['end_date'] =  pd.to_datetime(X2016_geo['end_date'], format='%Y-%m-%d %H:%M:%S')
```

```python
# Create day and hour columns for start_date and end_date columns.
X2016_geo['start_day'] = X2016_geo['start_date'].dt.date
X2016_geo['start_hour'] = X2016_geo['start_date'].dt.hour
X2016_geo['end_day'] = X2016_geo['end_date'].dt.date
X2016_geo['end_hour'] = X2016_geo['end_date'].dt.hour
```

```python
# Grouping by the day, hour and station to get number of journeys per hour for each day for each station.
starting = X2016_geo.groupby([X2016_geo['start_day'], X2016_geo['start_hour'], X2016_geo['start_station_name']]).size().unstack(fill_value=0).reset_index()
ending = X2016_geo.groupby([X2016_geo['end_day'], X2016_geo['end_hour'], X2016_geo['end_station_name']]).size().unstack(fill_value=0).reset_index()
```

```python
# Using melt to go from wide format to long format - getting the station names into a column.
starting_melt =  pd.melt(starting, id_vars = ['start_day', 'start_hour'], var_name = 'station', value_name = 'start_count')
ending_melt = pd.melt(ending, id_vars = ['end_day', 'end_hour'], var_name = 'station', value_name = 'end_count')
```

```python
# The ending melted dataframe includes values from 2017 as there are lots of journeys which started in 2016
# and ended in 2017!

first = pd.to_datetime('2017-01-01').date()
second = pd.to_datetime('2017-01-02').date()
third = pd.to_datetime('2017-01-03').date()
ending_melt = ending_melt[ending_melt.end_day != first]
ending_melt = ending_melt[ending_melt.end_day != second]
ending_melt = ending_melt[ending_melt.end_day != third]

#Reset index to allow for simple concatenation
ending_melt.reset_index(drop = True, inplace = True)
```

```python
# Check the dataframes are now both the same size. Yes! i.e. 366 days x 24 hours x 768 stations
print 'starting_melt shape:', starting_melt.shape
print 'ending_melt shape:', ending_melt.shape
```
